# Tensorboard

````python
from torch.utils.tensorboard import SummaryWriter
import cv2
from PIL import Image
import numpy as np
````

<br>

### SummaryWriter类

SummaryWriter类用于向tensorboard中写入信息

在实例化时，需要指定信息的存储目录，在写入信息后，会在项目目录下创建指定的目录

````python
writer = SummaryWriter('logs')    # 指定存储信息的目录
````

<br>

### add_scalar参数说明

- tag，绘制操作所在的画布名称，没有的话就新建画布

- scalar_value，y值

- global_step，x值

````python
for i in range(100):
    writer.add_scalar('y = x^2', i * i, i)
````

### 图像的数据结构

图像一般由三维数组组成，三个维度分别如下：

- H，表示高

- W，表示宽

- C，表示通道

H 与 W 可以确定图像的一个像素点，C 用于确定该像素点的色彩

<br>

### add_image参数说明

- tag，绘制操作所在的画布名称，没有的话就新建画布

- img_tensor，在该x值处的图像信息（要求为 Tensor 类型或 numpy.ndarray 类型）

- global_step，x值

- dataformats，指定图像数据的格式，也就是 CHW 三个维度的顺序（当图像是 Tensor 类型是，不需要指定）

````python
writer.add_image('image-change', np.array(img1), 1, dataformats='HWC')   # 不支持 PIL.Image 类型
writer.add_image('image-change', img2, 2, dataformats='HWC')
````

<br>

数据全部写入后，需要回收 writer 实例

````python
writer.close()
````

### 数据重写

tensorboard中对于同一个画布上的同一个 x 值上的 y 值的多次写入并不是覆盖写入

也就是说一个 global_step 可以对应多个 scalar_value

当发现绘制图像不是需求时，可以通过以下方式重新写入数据

- 1.删除项目目录下的存储 tensorboard 数据的目录

- 2.重新运行代码写入数据

- 3.命令行重新输入查看命令观测
