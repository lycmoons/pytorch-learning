# Transforms

transforms 是一个工具箱

其内有很多用于进行图像数据处理的工具类

这些工具类都是通过实例化后，用实例类调用call魔法方法来进行图像的处理

````python
from torchvision import transforms
from PIL import Image
import cv2
import matplotlib.pyplot as plt
````

````python
image_path1 = 'C:\\Users\\Administrator\\Desktop\\pytorch_learning\\dataset\\train\\ant\\0013035.jpg'
image_path2 = 'C:\\Users\\Administrator\\Desktop\\pytorch_learning\\dataset\\train\\bee\\16838648_415acd9e3f.jpg'
````

````python
image1 = Image.open(image_path1)
image2 = cv2.imread(image_path2)
type(image1), type(image2)   # 两种读取图像的方式得到的数据类型不同
````

<br>

### ToTensor类

ToTensor 类可以将 PIL Image 类型和 numpy.ndarray 类型的图像数据转为 Tensor 类型

````python
tentor_trans = transforms.ToTensor()
tensor_img1 = tentor_trans(image1)
tensor_img2 = tentor_trans(image2)
type(tensor_img1), type(tensor_img2)
````

<br>

### Resize类

对图像进行尺寸上的拉伸（不是裁剪）

````python
resize_trans = transforms.Resize((300, 300))  # 通过实例化参数创建个性化的工具
img1_resized = resize_trans(tensor_img1)
````

````python
transforms.ToPILImage()(tensor_img1)
````

````python
transforms.ToPILImage()(img1_resized)
````

<br>

### Compose类

使用多个 transforms 工具实例类的列表来进行实例化

通过这个 Compose 实例类调用 call 魔法方法后，就是顺序的执行其中的每一个图像处理

````python
trans_compose = transforms.Compose([tentor_trans, resize_trans])
img2_resized = trans_compose(image2)
````

````python
transforms.ToPILImage()(image2)
````

````python
transforms.ToPILImage()(img2_resized)
````
